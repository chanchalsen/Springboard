```python
"""Call logging within IT department of company X

In the past year, the IT department has let a couple of people go. 
The remaining employees had to take up their tasks. 
The hiring manager would like to understand what impact the departure of the individuals has had over the past year. 
He/she then plots the monthly volume of calls logged and resolved over the past calendar year. 
From the plot he/she sees some evidence of the teams productivity suffering from being short-staffed and 
now wants to turn this quick-and-dirty visual into a story to support his/her hiring request."""
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

sns.set_style()

df2 = pd.DataFrame({'Month': ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September',
                              'October', 'November', 'December'],
                    'Ticket_Volume_Received': [160, 184, 241, 149, 180, 161, 132, 202, 160, 139, 149, 177],
                    'Ticket_Volume_Processed': [160, 184, 237, 148, 181, 150, 123, 156, 126, 104, 124, 140]})
names = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November',
         'December']
names2 = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']


fig = plt.figure(figsize=(15,8))
ax = fig.add_subplot(111)
# Variables that define subsets of the data, which will be drawn on separate facets in the grid.
sns.barplot(x=df2.index,y=df2['Ticket_Volume_Received'],data=df2,label = 'Calls Logged',color='#4848FF',ax=ax)
sns.barplot(x=df2.index,y=df2['Ticket_Volume_Processed'],data=df2,label = 'Calls Resolved',color='#FF6800',ax=ax)
plt.title('Monthly Volume of IT calls Logged and Resolved',fontsize=14,ha='right')
# Put a legend below current axis
plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.08),fancybox=True, shadow=False, ncol=5,fontsize=12)
plt.grid(axis='y')
plt.ylabel('')
plt.ylim(0,300)
ax.set(xticks=range(0, 11), xticklabels=list(names))
plt.show()
```

```python
fig = plt.figure(figsize=(15,8))
ax = fig.add_subplot(111)
# Variables that define subsets of the data, which will be drawn on separate facets in the grid.
sns.lineplot(x=df2.index,y=df2['Ticket_Volume_Received'],data=df2,marker='o',markersize=10,lw=4,label = 'Calls Logged',color='#4848FF',ax=ax)
sns.lineplot(x=df2.index,y=df2['Ticket_Volume_Processed'],data=df2,marker='s',markersize=10,lw=4,label = 'Calls Resolved',color='#FF6800',ax=ax)
plt.title('Monthly Volume of IT calls Logged and Resolved',fontsize=14,ha='right')
# Put a legend below current axis
plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.08),fancybox=True, shadow=False, ncol=5,fontsize=12)
plt.grid(axis='y')
plt.ylabel('')
plt.ylim(0,300)
ax.set(xticks=range(0, 11), xticklabels=list(names))
plt.show()
```

```python
fig = plt.figure(figsize=(15,8))
ax = fig.add_subplot(111)
# Variables that define subsets of the data, which will be drawn on separate facets in the grid.
sns.lineplot(df2.index,y=df2['Ticket_Volume_Received'],marker='o',markersize=10,data=df2,lw=4,color='#4848FF',ax=ax)
sns.lineplot(df2.index,y=df2['Ticket_Volume_Processed'],marker='s',markersize=10,data=df2,lw=4,color='#FF6800',ax=ax)
plt.title('Monthly Volume of IT calls Logged and Resolved',fontsize=18,ha='right',color='#616161')
sns.despine()
# Change the colors of the left and bottom borders (fade into the background)
ax.spines['left'].set_color('lightgray')
ax.spines['bottom'].set_color('lightgray')
ax.text(12.6, 180, 'Logged', ha='right', va='top', fontsize=22,color='#4848FF')
ax.text(12.8, 145, 'Resolved', ha='right', va='top', fontsize=22,color='#FF6800')
plt.ylabel('')
plt.ylim(0,300)
ax.set(xticks=range(0, 11), xticklabels=list(names2))
plt.ylabel('Volume',fontsize=16,color='#616161')
fig.text(
        0.05, 0.05,
        'Adapted from: Storytelling with Data by Cole Nussbaumer Knaflic',
        ha='left',fontsize=12)
plt.show()
```
